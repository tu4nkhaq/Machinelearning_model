* II.Sử dụng phương pháp K-fold cross validation
Sử dụng K-fold cross validation kết hợp với 1 trong số các phương pháp hồi quy: hoặc Hồi
quy tuyến tính hoặc LASSO hoặc RIDGE, để thực hiện các yêu cầu sau cho từng bài toán
trong mục I:
* 1) Chia tập huấn luyện thành k phần.
* 2) Sử dụng các phương pháp hồi quy đã chọn (hoặc Hồi quy tuyến tính hoặc Lasso hoặc
Ridge) và k-fold cross validation để lựa chọn mô hình tốt nhất cho bài toán (mô hình
có tổng lỗi trên tập train và tập validation là nhỏ nhất).
* Gợi ý:
* Bước 1: Chia toàn bộ tập dữ liệu huấn luyện (X_train, y_train) thành k phần (phương
pháp k-fold cross validation).
* Bước 2: Chọn ngẫu nhiên k-1 phần làm train data, 1 phần còn lại làm validation data.
Sử dụng phương pháp học máy đã lựa chọn để huấn luyện mô hình trên tập train data,
tính tổng lỗi của mô hình trên tập train data và tập test data. Bước 2 này được làm k
lần, để chọn mô hình có tổng lỗi nhỏ nhất.
* Bước 3: Chọn mô hình có (train error + validation error) là nhỏ nhất.Ví dụ:
* Bước 1: Chia tập dữ liệu huấn luyện thành 3 tập A, B, C.
* Bước 2: Huấn luyện mô hình trên tập train data, tính tổng lỗi của mô hình trên tập train
data và validation data.
* - Lần 1: training data: A, B; validation data: C.
* - Lần 2: training data: A, C; validation data: B.
* - Lần 3: training data: B, C; validation data: A.
Trong 3 lần trên, lần nào có (train error + validation error) nhỏ nhất thì mô hình huấn
luyện của lần đó được chọn làm mô hình dự đoán cho dữ liệu mới.
* 3) Dùng tập dữ liệu kiểm tra (test data) để đánh giá chất lượng mô hình bằng các độ đo:
R2, NSE, MAE, RMSE

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
import pandas as pd
import numpy as np
import math

#hàm chuẩn hóa các chuỗi kí tự có trong dữ liệu
def dataEncoder(dataStr):
	dataConvert = dataStr
	for i, j in enumerate(dataConvert):
		for k in range(0,len(dataConvert[0])):
			if(j[k]=="none"):
				j[k] = 0
			elif(j[k]=="low"):
				j[k] = 1
			elif(j[k]=="med"):
				j[k] = 2
			elif(j[k]=="high"):
				j[k] = 3
	return dataConvert # trả về dữ liệu đã được chuẩn hóa

#hàm tính lỗi
def Error(Y, Y_pred):
	Array = [] #mảng lưu giá trị chênh lệnh gữa Y dự đoán vs Y thực tế
	for i in range(0,len(Y)):
		Array.append(np.abs(np.array(Y[i:i+1]) - np.array(Y_pred[i:i+1])))
	return np.mean(Array) #trả về trung bình giá trị chênh lệch

#đọc dữ liệu từ file
Data = pd.read_csv('Data.csv')

#chia dữ liệu 70% để train 30% để test, có xáo trộn dữ liệu
dataTrain, dataTest = train_test_split(dataEncoder(np.array(Data.values)), test_size = 0.3, shuffle = True)

#Khai báo KFold chia dữ liệu tập train thành 5 phần
KF = KFold(n_splits = 5, random_state = None)

#tìm kiếm mô hình huấn luyện tốt nhất
modelIndex  = 1
MAX = math.inf
for train_index, test_index in KF.split(dataTrain):
	#cắt dữ liệu vào các biến tương ứng (X chứa các thuộc tính | Y chứa nhãn)
	X_train, X_test = dataTrain[train_index, :6], dataTrain[test_index, :6]
	Y_train, Y_test = dataTrain[train_index, 6], dataTrain[test_index, 6]

	#huấn luyện trên mô hình đã được chọn
	Lr = LinearRegression().fit(X_train, Y_train)

	#dự đoán kết quả dựa trên mô hình đã chọn
	Y_pred_train = Lr.predict(X_train)
	Y_pred_test = Lr.predict(X_test)

	#tính tổng trung bình lỗi trên mô hình đã chọn
	SUM = Error(Y_train, Y_pred_train) + Error(Y_test, Y_pred_test)

	#xét điều kiện để chọn mô hình tốt nhất
	if(SUM<MAX):
		MAX = SUM
		bestModelIndex = modelIndex
		Regr = Lr.fit(X_train, Y_train)
	modelIndex = modelIndex + 1

#bắt đầu dự đoán dựa trên mô hình tốt nhất
Y_predict = Regr.predict(dataTest[:, :6])

print(f"Vector trọng số W = {Lr.coef_}\nHệ số tự do W0 = {Lr.intercept_}")
print(f"Mức độ phù hợp của mô hình = {Lr.score(X_test,Y_test)}")
```

```python
import numpy as np
#Tính w của Hồi quy tuyến tính không có hệ số tự do bias
X_train = np.array([[60, 2, 10], [40, 2, 5], [100, 3, 7], [80, 4, 6]]).T
y_train = np.array([10, 12, 20, 15]).T
x_test = np.array([[50, 2, 8]]).T

w = np.linalg.pinv(X_train@X_train.T)@X_train@y_train
y_test_hat = x_test.T@w
print('Tính theo công thức, Hồi quy tuyến tính không có bias:')
print ('w', w)
print('Giá trị dự đoán mẫu mới: ', y_test_hat)
print('Giá trị dự đoán tập huấn luyện: ', X_train.T@w)

#Tính w của Hồi quy tuyến tính có hệ số bias
X_train = np.array([[60, 2, 10], [40, 2, 5], [100, 3, 7], [80, 4, 6]]).T
one = np.ones((1, X_train.shape[1]))
X_train = np.concatenate((one, X_train), axis=0)
y_train = np.array([10, 12, 20, 15]).T

X_test = np.array([[50, 2, 8]]).T
one = np.ones((1, X_test.shape[1]))
X_test = np.concatenate((one, X_test), axis=0)

w = (np.linalg.pinv(X_train@X_train.T))@X_train@y_train
y_test_hat = X_test.T@w
print('Tính theo công thức, Hồi quy tuyến tính có bias')
print ('Giá trị theo công thức của w:', w)
print('Giá trị dự đoán mẫu mới bằng công thức: ', y_test_hat)
print('Giá trị dự đoán tập huấn luyện bằng công thức: ', X_train.T@w)

# Tính w bằng LinearRegression của thư viện sklear
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[60, 2, 10], [40, 2, 5], [100, 3, 7], [80, 4, 6]])
y = np.array([10, 12, 20, 15])

model = LinearRegression()
model.fit(X, y)
w0=model.intercept_
w1=model.coef_
print('Tính w bằng thư viện scikit-learn:')
print(' w0 = ', w0, 'w_1 = ', w1)

X_test = np.array([[50, 2, 8]])
print('y_test_hat: ', model.predict(X_test))
print('y_train_hat: ', model.predict(X))




```
