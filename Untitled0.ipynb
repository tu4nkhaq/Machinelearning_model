```python
from math import *

def grad(x):
    return 2*x

def GD(x, eta, k):
    i=0
    for i in range(k):
      x -= eta*grad(x)
      if grad(x) < 0.0001:
           break
    return x

x = GD(0, 0.01, 100)
print("Giá trị để f'(x)~0 là x = ", x)
print("f'({}) = ".format(x), grad(x))
```
